```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('./train.csv')
train.head()
```

```python
def decode_pixels(encode_pixels, img_shape=(1400, 2100)):
    starts = np.array(list(map(int, encode_pixels.split()[::2])))
    lens = np.array(list(map(int, encode_pixels.split()[1::2])))
    ###
    starts -= 1
    ends = starts + lens

    mask = np.zeros(1400*2100, dtype=np.uint8)
    for i, j in zip(starts, ends):
        mask[i:j] = 1

    return mask.reshape(img_shape, order='F')
```

```python
image1 = plt.imread('./0011165.jpg')
plt.imshow(image1)
mask = decode_pixels(train.loc[0, 'EncodedPixels'])
plt.imshow(mask, alpha=0.5, cmap='gray')
```
